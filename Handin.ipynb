### 1. The union of all the tracks with genreid 18 and 20
 - We get 39 destinct tracks when we union tracks with genreid 20 and 18. Track can only have on genreid.
- All tracks returned are distinct. 

````python
%sql SELECT name FROM genre WHERE genreid=18 OR genreid=20;
````

````python
%sql(SELECT name, composer, trackid FROM track  WHERE genreid=20 ) UNION  (SELECT name, composer, trackid FROM track  WHERE genreid=18 ) order by trackid;
````

 ### 2. The intersection of all the invoices that are cheaper than 10 dollars and the invoices that are more expensive than 5 dollars
 - 115 distinct rows are returned. Because the intersect is returned what we get is the set of invoices with total between the intavl > 5 and < 10.

````python
%sql(SELECT * FROM invoice  WHERE total<10) intersect (SELECT * FROM invoice WHERE total > 5) ORDER BY total;
````

 ### 3. The set of all customers from USA, subtracted by the set of all customers with an email ending in 'yahoo.com'
- First I checked what the exact country code were for the query.
- Then I used except to return the set of customers from the USA except the set where those customers have an email ending with '%yahoo.com'
- We can see that 13 customers are from the USA and if we exclude those with the email ending the result returned is only 11.

````python
%sql SELECT country FROM customer WHERE country LIKE 'US%';
%sql SELECT * FROM customer WHERE country = 'USA';
%sql (SELECT * FROM customer WHERE country = 'USA') except (SELECT * FROM customer WHERE email LIKE '%yahoo.com');
````

 ### 4. The union of the set of all albums playing something by Mozart and the set of all albums playing something with Bach
- The albumid and composer exist in the track table. We select union the sets that has the composer Bach or Mozart and select the albumid. 
- Then we select the set in albums with the matching set of albumids.

````python
%sql SELECT * FROM track WHERE composer LIKE '%Bach';
%sql SELECT * FROM track WHERE composer LIKE '%Mozart';
````

````python
%sql (SELECT albumid FROM track WHERE composer LIKE '%Bach' order by albumid) union (SELECT albumid FROM track WHERE composer LIKE '%Mozart' order by albumid);
````

````python
%sql (SELECT album.albumid,title,artistid FROM album INNER JOIN track ON album.albumid=track.albumid WHERE composer LIKE '%Bach' ) union (SELECT album.albumid,title,artistid FROM album INNER JOIN track ON album.albumid=track.albumid WHERE composer LIKE '%Mozart');
````
